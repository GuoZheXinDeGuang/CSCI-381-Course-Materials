```python
#### (a) Your code here
import numpy as np
import pandas as pd
df=pd.read_csv("bank.csv",sep=";")
df
```

Try to discover this data:

```python
#### (c) Your code here
df=df.replace("unknown",np.nan)
print("The missing value in each column:\n",np.sum(df.isna(),axis=0))
print("Total missing value:",np.sum(np.sum(df.isna(),axis=1)))
```

```python
#### (d) Your code here
for i in df.columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
df
```

```python
#### (e) Your code here
for y in (df.drop(['age','balance'],axis=1).columns):
    print(df.groupby(y).size(),'\n')
```

```python
#### (f) Your code here
age=df.age
balance=df.balance
print("For age:\nthe mean is",age.mean(),"\nthe max is", age.max(),"\nthe min is",age.min())
print("For balance:\nthe mean is",balance.mean(),"\nthe max is", balance.max(),"\nthe min is",balance.min())
```

## Question-4

- a. Find the number of people of 'secondary' education level.
- b. Compare between the average of the balance of married and single people
- c. Find the mean and the standard deviation of the balance of each education level.
- d. Find the mean and the standard deviation of the balance of each job.
- e. What is the average balance for different education groups and marital status?
- f. Compare between the number of people who had loan (married vs single)

- h. Which job has the highest number of loans?


```python
#### (a) Your code here
print("The number of people in secondary level is:",len(df.loc[df.education=='secondary']))
```

```python
#### (b) Your code here
df[(df['marital']=='single') | (df['marital']=='married')][['marital','balance']].groupby('marital').agg(['mean'])
```

```python
#### (c) Your code here
df[['education','balance']].groupby('education').agg(['mean','std'])
```

```python
#### (d) Your code here
df[['job','balance']].groupby('job').agg(['mean','std'])
```

```python
#### (e) Your code here
df[['education','marital','balance']].groupby(['education','marital']).agg(['mean'])
```

```python
#### (f) Your code here
def num_loan(x):
    return np.sum(x=='yes')
df[(df['marital']=='single') | (df['marital']=='married')][['marital','loan']].groupby('marital').agg([num_loan])
```

```python
#### (g) Your code here
jobs=df['job'].unique()
job_balance=[]
for i in list(jobs):
    job_balance.append((i,df.loc[df.job==i]['balance'].mean()))
print(sorted(job_balance,key=lambda n:n[1])[0][0])
```

```python
#### (h) Your code here
jobs=df['job'].unique()
job_loan=[]
for i in list(jobs):
    count = len(df.loc[(df['job'] == i) & (df['loan'] == 'yes')])
    job_loan.append((i, count))
print(sorted(job_loan,key=lambda n:n[1])[-1][0])
```
